```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Laptop_data.xlsx')
df.head()
```

```python
df['Price']=df['Price'].str.replace('₹','')
df['Original Price']=df['Original Price'].str.replace('₹','')
df['Price']=df['Price'].str.replace(',','')
df['Original Price']=df['Original Price'].str.replace(',','')
df['Discount']=df['Discount'].str.replace('% off','')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.astype({'Discount':'float'})
```

```python
df = df.astype({'Price':'int'})
df = df.astype({'Original Price':'float'})
```

```python
df.info()
```

```python
df.hist(figsize=(10,10))
```

```python
sns.regplot(df['Discount'],df['Ratings'])
```

Ratings are increasing as the discount is increasing.

```python
sns.regplot(df['Price'],df['Ratings'])
```

Ratings are decreasing as the price of laptops is increasing.
it seems rating is high when price is between 50,000 to 20,00,00 & ratings is getting low when price is between 20,00,00 to 40,00,00.

```python
sns.regplot(df['Original Price'],df['Ratings'])
```
